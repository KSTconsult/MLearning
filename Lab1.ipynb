### kNN

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors = 10)
knn.fit(X_train, y_train)
```

```python

y_train_predict = knn.predict(X_train)
y_test_predict = knn.predict(X_test)

err_train = np.mean(y_train != y_train_predict)
err_test  = np.mean(y_test  != y_test_predict)

print(err_train, err_test)
```

```python
from sklearn.metrics import confusion_matrix

print(confusion_matrix(y_test, y_test_predict))
```

Из матрицы видно, что ошибок очень мало. Метод 11 раз ошибся, говоря, что вино белое, хотя оно было красным, и всего 5 раз сказал, что белое вино - красное.

###### Подберем наилучшее число соседей

```python
from sklearn.model_selection import GridSearchCV
nnb = [1, 3, 5, 10, 15, 20, 25, 35, 45, 55]
knn = KNeighborsClassifier()
grid = GridSearchCV(knn, param_grid = {'n_neighbors': nnb}, cv=10) #Подборка параметров
grid.fit(X_train, y_train)

best_cv_err = 1 - grid.best_score_
best_n_neighbors = grid.best_estimator_.n_neighbors
print("Наименьшая ошибка " + str(best_cv_err))
print("Оптимальное число соседей соседей " + str(best_n_neighbors))
```

```python
knn = KNeighborsClassifier(n_neighbors = best_n_neighbors).fit(X_train, y_train)

err_train = np.mean(y_train != knn.predict(X_train))
err_test  = np.mean(y_test  != knn.predict(X_test))

print("train error: " + str(err_train)) 
print("test error: " + str(err_test))
```

### PCA

###### Применяем PCA при этом сокращая размерность признаков до двух

```python
pca = PCA(n_components = 2)
X_PCA = pca.fit_transform(X)
```

###### Результат

```python
X_PCA
```

```python
plt.scatter(X_PCA[:,0], X_PCA[:,1], alpha = .2)

plt.xlabel('v1')
plt.ylabel('v2')
plt.title('2 components after PCA')
```

###### Применяем kNN с оптимальным числом соседей равным 5

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_PCA, y, test_size = 0.3, random_state = 42)

N_train, _ = X_train.shape 
N_test,  _ = X_test.shape 

print(N_train, N_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 5).fit(X_train, y_train)

err_train = np.mean(y_train != knn.predict(X_train))
err_test  = np.mean(y_test  != knn.predict(X_test))

print("train error: " + str(err_train)) 
print("test error: " + str(err_test))
```

###### Построим график зависимости ошибки от главных компонент

```python
comp = range(1, 12, 1)
err_train_mas = []
err_test_mas = []
for temp in comp:
    pca = PCA(n_components = temp)
    X_PCA = pca.fit_transform(X)
    X_train, X_test, y_train, y_test = train_test_split(X_PCA, y, test_size = 0.3, random_state = 42)
    knn = KNeighborsClassifier(n_neighbors = 5).fit(X_train, y_train)
    err_train_mas.append(np.mean(knn.predict(X_train) != y_train))
    err_test_mas.append(np.mean(knn.predict(X_test) != y_test))
plt.plot(comp, err_train_mas, 'r-o', label = 'train error')
plt.plot(comp, err_test_mas,  'b-o', label = 'test error') 
plt.title('PCA')
plt.xlabel('components')
plt.ylabel('error')
plt.legend(loc = 2)
```

###### Из графика видно, что самая маленькая ошибка достигается при использовании 11 компонент

###### Найдем минимальную ошибку

```python
np.min(err_test_mas)
```

###### Индекс оптимального количества главных компонент

```python
np.argmin(err_test_mas)
```

Минимальная ошибка: 0.008251676121712223

Оптимальное количество компонент: 11

Вывод: метод главных компонент позволил снизить ошибку на 0.0005 по сравнению с kNN при использовании 5 соседей.

### Нейронные сети

Разбиваем данные на обучающую и тестовую выборки

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.33, random_state = 7)
N_test, _ = X_test.shape
```

Обучаем классификатор

```python
from sklearn.neural_network import MLPClassifier
```

Обучение классификатора в виде многослойного персептрона. Активационная функция: сигмоида. Один скрытый слой, со 100 нейронами.

```python
mlp_model = MLPClassifier(hidden_layer_sizes = (100,), solver = 'lbfgs', activation = 'logistic', random_state = 42)
mlp_model.fit(X_train, y_train)
```

```python
y_train_pred = mlp_model.predict(X_train)
y_test_pred = mlp_model.predict(X_test)


err_train = np.mean(y_train != y_train_pred)
err_test = np.mean(y_test != y_test_pred)

print("ошибка на обучающей выборке: " + str(err_train)) 
print("ошибка на тестовой выборке: " + str(err_test))
```

Попробуем использовать двухслойную полносвязную сеть.

```python
mlp_model = MLPClassifier(hidden_layer_sizes = (60, 10), solver = 'lbfgs', activation = 'logistic', random_state = 42)
mlp_model.fit(X_train, y_train)
```

```python
y_train_pred = mlp_model.predict(X_train)
y_test_pred = mlp_model.predict(X_test)


err_train = np.mean(y_train != y_train_pred)
err_test = np.mean(y_test != y_test_pred)

print("ошибка на обучающей выборке: " + str(err_train)) 
print("ошибка на тестовой выборке: " + str(err_test))
```

### Подбор гиперпараметров

Подберем оптимальное значение параметра регуляризации alpha

```python
alpha_massive = np.logspace(-3, 3, 21)
mlp = MLPClassifier(hidden_layer_sizes = (100,), solver='lbfgs', activation = 'logistic')

grid = GridSearchCV(mlp, param_grid={"alpha": alpha_massive}, cv=5)
grid.fit(X_train, y_train)

best_cv_err = 1 - grid.best_score_
best_alpha = grid.best_estimator_.alpha
print("Меньшая ошибка " + str(best_cv_err))
print("Лучший параметр alpha " + str(best_alpha))
```

```python
mlp = MLPClassifier(alpha = best_alpha)
mlp.fit(X_train, y_train)
err_train = np.mean(y_train != mlp.predict(X_train)) #ошибка на обуч. выборке
err_test  = np.mean(y_test  != mlp.predict(X_test)) #ошибка на тесте

print("ошибка на обучающей выборке при лучшей alpha: " + str(err_train)) 
print("ошибка на тестовой выборке при лучшей alpha: " + str(err_test))
```

```python
alpha_arr = np.logspace(-3, 2, 21)
test_err = []
train_err = []
for alpha in alpha_arr:
    mlp_model = MLPClassifier(alpha = alpha, hidden_layer_sizes = (100,), 
                              solver = 'lbfgs', activation = 'logistic', random_state = 42)
    mlp_model.fit(X_train, y_train)

    y_train_pred = mlp_model.predict(X_train)
    y_test_pred = mlp_model.predict(X_test)
    train_err.append(np.mean(y_train != y_train_pred))
    test_err.append(np.mean(y_test != y_test_pred))
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.semilogx(alpha_arr, train_err, 'b-o', label = 'train')
plt.semilogx(alpha_arr, test_err, 'r-o', label = 'test')
plt.xlim([np.max(alpha_arr), np.min(alpha_arr)])
plt.title('Error vs. alpha')
plt.xlabel('alpha')
plt.ylabel('error')
plt.legend()
```

Меняем число нейронов в скрытом слое

```python
neuron_massive = np.linspace(10, 100, 60, dtype = int)
mlp = MLPClassifier(alpha = best_alpha, solver='lbfgs', activation = 'logistic')

grid = GridSearchCV(mlp, param_grid={"hidden_layer_sizes": neuron_massive}, cv=5)
grid.fit(X_train, y_train)

best_cv_err = 1 - grid.best_score_
best_neuron = grid.best_estimator_.hidden_layer_sizes
print("Меньшая ошибка " + str(best_cv_err))
print("Лучший параметр alpha " + str(best_neuron))
```

```python
mlp = MLPClassifier(hidden_layer_sizes = best_neuron, alpha = best_alpha, 
                    solver='lbfgs', activation = 'logistic')
mlp.fit(X_train, y_train)
err_train = np.mean(y_train != mlp.predict(X_train)) #ошибка на обуч. выборке
err_test  = np.mean(y_test  != mlp.predict(X_test)) #ошибка на тесте

print("ошибка на обучающей выборке при лучшей alpha: " + str(err_train)) 
print("ошибка на тестовой выборке при лучшей alpha: " + str(err_test))
```

```python
err_train_mas = []
err_test_mas = []
for cur in neuron_massive:
    model = MLPClassifier(alpha = best_alpha, hidden_layer_sizes = cur, 
                          solver='lbfgs', activation = 'logistic')
    model.fit(X_train, y_train)
    
    err_train_mas.append(np.mean(model.predict(X_train) != y_train))
    err_test_mas.append(np.mean(model.predict(X_test) != y_test))
    
plt.semilogx(neuron_massive, err_train_mas, 'r-o', label = 'Train error')
plt.semilogx(neuron_massive, err_test_mas,  'b-o', label = 'Test error') 
plt.xlim([np.max(neuron_massive), np.min(neuron_massive)])

plt.title('MLP neurons/error')
plt.xlabel('neurons')
plt.ylabel('error')
plt.legend(loc = 'best')
```

На графиках ошибка на тестовой выборке при уменьшении alpha сначала убывает, достигает локального минимума, а с последующем уменьшением параметра, ошибка немного начинает расти (начинается мизерное переобучение). А так в целом видно видно, что алгоритм дал высокую точность на обеих выборках. Полученные ошибки пренебрежительно малы.

### Random forest

```python
from sklearn import ensemble
rf = ensemble.RandomForestClassifier(n_estimators = 100)
rf.fit(X_train, y_train)
```

Считаем среднюю ошибку

```python
err_train = np.mean(y_train != rf.predict(X_train))
err_test  = np.mean(y_test  != rf.predict(X_test))

print(err_train, err_test)
```

Построим график зависимости ошибки на обучающей и тестовой выборках,а также out-of-bag ошибки (oob-ошибки) от количества деревьев.

```python
ran_tree = range(25, 2500, 100)
err_train_mas = []
err_test_mas = []
err_oob = []
for t_cur in ran_tree:
    mod_t = ensemble.RandomForestClassifier(n_estimators = t_cur, oob_score=True)
    mod_t.fit(X_train, y_train)
    oob_error = 1 - mod_t.oob_score_
    err_train_mas.append(np.mean(mod_t.predict(X_train) != y_train))
    err_test_mas.append(np.mean(mod_t.predict(X_test) != y_test))
    err_oob.append(oob_error)
plt.plot(ran_tree, err_train_mas, 'r-o', label = 'Train error')
plt.plot(ran_tree, err_test_mas,  'b-o', label = 'Test error') 
plt.plot(ran_tree, err_oob,  'g-o', label = 'oob error')    
plt.title('Random Forest')
plt.xlabel('number of estimators')
plt.ylabel('error')
plt.legend(loc = 'best')
```

Ошбики составляют менее 1%, все хорошо. Наблюдается мизерное переобучение, которое несущественно.

### Gradient Boosted Trees

```python
gbt = ensemble.GradientBoostingClassifier(n_estimators = 1000)
gbt.fit(X_train, y_train)

err_train = np.mean(y_train != gbt.predict(X_train))
err_test = np.mean(y_test != gbt.predict(X_test))

print("ошибка на обучающей выборке: " + str(err_train)) 
print("ошибка на тестовой выборке: " + str(err_test))
```

Построим график зависимости ошибкок на обучающей и тестовой выборках от количества деревьев.

```python
ran_tree = range(1, 1500, 100)
err_train_mas = []
err_test_mas = []
err_oob = []
for t_cur in ran_tree:
    mod_t = ensemble.GradientBoostingClassifier(n_estimators = t_cur)
    mod_t.fit(X_train, y_train)
    err_train_mas.append(np.mean(mod_t.predict(X_train) != y_train))
    err_test_mas.append(np.mean(mod_t.predict(X_test) != y_test))
    err_oob.append(np.mean(mod_t.predict(X_test) != y_test))
plt.plot(ran_tree, err_train_mas, 'r-o', label = 'Train error')
plt.plot(ran_tree, err_test_mas,  'b-o', label = 'Test error')
plt.plot(ran_tree, err_oob,  'b-o', label = 'oob error')    

plt.legend(loc = 2)
```

Ошибки колеблются в районе одного процента, переобучения нет.

Все алгоритмы дают отличные результаты. Ошибка на тестовой выборке меньше 1%.
